```python
import wfdb
import pandas as pd

# Load record (NO .dat or .hea extension)
record = wfdb.rdrecord('/content/00001_lr')

# Show leads
print("Leads:", record.sig_name)

# Convert to DataFrame
df = pd.DataFrame(record.p_signal, columns=record.sig_name)

# Show preview
print(df.head())

# Save to CSV
output_path = "/content/00001_lr.csv"
df.to_csv(output_path, index=False)

print("Saved CSV at:", output_path)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fs = record.fs  # sampling frequency

start_time = 2   # seconds
end_time = 5     # seconds

start_idx = int(start_time * fs)
end_idx = int(end_time * fs)

plt.figure(figsize=(12,4))
plt.plot(df['II'][start_idx:end_idx])

plt.title("ECG Lead II (2s to 5s window)")
plt.xlabel("Samples")
plt.ylabel("Amplitude (mV)")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fs = record.fs  # sampling frequency

start_time = 2   # seconds
end_time = 5     # seconds

start_idx = int(start_time * fs)
end_idx = int(end_time * fs)

leads = df.columns  # all 12 leads

plt.figure(figsize=(15, 12))

for i, lead in enumerate(leads):
    plt.subplot(6, 2, i + 1)
    plt.plot(df[lead][start_idx:end_idx])
    plt.title(lead)
    plt.ylabel("mV")
    plt.grid()

plt.suptitle("12-Lead ECG (2s to 5s window)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

leads = df.columns

fig = go.Figure()

# Add all leads (only one visible at a time for clarity)
for i, lead in enumerate(leads):
    fig.add_trace(
        go.Scatter(
            y=df[lead],
            mode='lines',
            name=lead,
            visible=(lead == 'II')  # default visible lead
        )
    )

# Dropdown menu to switch leads
buttons = []

for i, lead in enumerate(leads):
    visibility = [False] * len(leads)
    visibility[i] = True

    buttons.append(dict(
        label=lead,
        method="update",
        args=[{"visible": visibility},
              {"title": f"ECG Lead: {lead} (Zoom enabled)"}]
    ))

fig.update_layout(
    title="Interactive 12-Lead ECG (Zoom In/Out Enabled)",
    xaxis_title="Samples",
    yaxis_title="Amplitude (mV)",
    updatemenus=[dict(
        buttons=buttons,
        direction="down",
        showactive=True
    )]
)

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 1. SYSTEM MODEL (UNFORCED)
# ==============================

A = np.array([[0.9, 0.1],
              [0.0, 0.95]])

C = np.array([[1.0, 0.0]])

x_true = np.array([[1.0],
                   [0.5]])

x_est = np.array([[0.0],
                  [0.0]])

P = np.eye(2)

Q = np.zeros((2, 2))

# ⚠️ Small numerical stabilizer (VERY IMPORTANT)
R = np.array([[1e-9]])

# ==============================
# 2. STORAGE
# ==============================

steps = 50
true_history = []
est_history = []
error_history = []

# ==============================
# 3. KALMAN FILTER LOOP
# ==============================

for k in range(steps):

    # True system (no noise)
    x_true = A @ x_true
    y = C @ x_true

    # Prediction
    x_pred = A @ x_est
    P_pred = A @ P @ A.T + Q

    # Innovation covariance
    S = C @ P_pred @ C.T + R

    # Kalman gain (stable inverse)
    K = P_pred @ C.T @ np.linalg.inv(S)

    # Update
    x_est = x_pred + K @ (y - C @ x_pred)
    P = (np.eye(2) - K @ C) @ P_pred

    # Store
    true_history.append(x_true[0, 0])
    est_history.append(x_est[0, 0])
    error_history.append(abs(x_true[0, 0] - x_est[0, 0]))

# ==============================
# 4. PLOTS
# ==============================

plt.figure(figsize=(12,5))
plt.plot(true_history, label="True State")
plt.plot(est_history, '--', label="Kalman Estimate")
plt.title("Kalman Filter - Stable Noiseless System")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12,4))
plt.plot(error_history, color='red')
plt.title("Estimation Error")
plt.grid()
plt.show()
```
